## customized surface

```python
import numpy as np
```

```python
x=np.arange(-5,6)
y=np.arange(-5,6)
```

```python
xv,yv=np.meshgrid(x,y)
```

```python
xv
```

```python
yv
```

```python
z=xv**2+yv**2
z
```

```python
surface=go.Surface(x=xv,y=yv,z=z)
fig=go.Figure(surface)
fig.show()
```
