```python
import plotly.graph_objects as go
import pandas as pd
```

```python
data=pd.read_csv('./data/3d-line1.csv')
data.head()
```

```python
line=go.Scatter(x=data['x'],y=data['y'])
#後面加上mode='markers'可以讓線不要連起來
fig=go.Figure(line)
fig.show()
```

```python
line=go.Scatter3d(x=data['x'],
                y=data['y'],
                z=data['z'])
fig=go.Figure(line)
fig.show()
```

```python
line=go.Scatter3d(x=data['x'],
                y=data['y'],
                z=data['z'],
                mode='lines')
#改成點用mode='markers'
fig=go.Figure(line)
fig.show()
```

```python
line=go.Scatter3d(x=data['x'],
                y=data['y'],
                z=data['z'],
                mode='markers',
                marker={'size':3,'color':'red'})
#改成點用mode='markers'
fig=go.Figure(line)
fig.show()
```
